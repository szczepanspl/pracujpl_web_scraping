```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_json("offers.json")
data

```

```python
#Checking if there are same offers
data["Offer Name"].value_counts()
```

```python
#Droping same offers
data = data.drop_duplicates(subset=["Offer Name", "Company"])
data

```

```python
# Most offers are released on February 1st
data["Date"].value_counts()
```

```python
#Cheking tags with skills that are needed for certain job
tag_lists = data["Tags"].dropna()
tags = []
for tag_list in tag_lists:
    for tag in tag_list:
        tags.append(tag)
tag_data = pd.Series(tags)
tag_data.head()
```

```python
# Picking only first 20 needed skills
most_needed_skills = tag_data.value_counts().head(20)
```

```python
# Ploting the bar chart
fig = plt.figure(figsize = (12, 4))
plt.bar(most_needed_skills.index, most_needed_skills)
plt.tick_params("x", labelrotation= 45)
plt.xlabel("Skills")
plt.ylabel("Number of offers")
plt.title("Most needed skills on pracuj.pl")
plt.show()
```
